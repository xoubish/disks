## Checking whether GAN results convolved with PSF generate galaxy images used in training

```python
import astropy.io.fits as pyfits
import matplotlib.pyplot as plt
import numpy as np
from astropy.convolution import convolve_fft as convolve

%matplotlib inline
```

```python
a = pyfits.getdata('test.fits')
psf = pyfits.getdata('psf_h.fits')
b = convolve(a,psf)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(a,origin='lower')
plt.text(10,50,'before PSF',color='y',size=15)
plt.colorbar()

plt.subplot(1,2,2)
b=np.tanh(b)
plt.imshow(b,origin='lower')
plt.text(10,50,'convolved with PSF',color='y',size=15)
plt.colorbar()
plt.tight_layout()
```

Confirming GAN is working fine. Now, need to add some high pass filtering, or change the cost function of GAN.
