# Fitting an IFU disk to measure parameters 

Galpak is an mcmc code which can be used to fit 3d IFU data of disks, and retreive parameters of the disk including fluxes, inclination, radii, maximum velocity, etc. . Currently, it does not measure shear but parameters can be added if needed. Here I just check the package on one MUSE cube.

```python
import numpy as np
import matplotlib.pyplot as plt
import galpak
```

```python
mydisk = galpak.DiskModel(flux_profile='gaussian', rotation_curve='isothermal')
```
